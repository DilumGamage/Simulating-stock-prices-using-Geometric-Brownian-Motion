```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
S0 = 934.74 #Netfilx Data
sigma = 0.293868679 #volatility of stock price
r = 0.045 # risk free rate
T = 30 / 252 #time duration measured in years
n = 30  # Number of steps
delta_t = T / n

# Calculate binomial model parameters
u = np.exp(sigma * np.sqrt(delta_t))
d = 1 / u
p = (np.exp(r * delta_t) - d) / (u - d)
```

```python
# Simulate 50 paths
num_paths = 50
paths = np.zeros((n + 1, num_paths))
paths[0] = S0

for i in range(1, n + 1):
    z = np.random.rand(num_paths)
    paths[i] = paths[i - 1] * np.where(z < p, u, d)

# Plot the paths
plt.figure(figsize=(10, 6))
plt.plot(paths)
plt.title("Simulated Binomial Paths for Netflix Stock")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```
